```python
import matplotlib.pyplot as plt
# Read the CSV file
df = pd.read_csv('customers_export.csv')

# Convert columns to numeric
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Total Orders'] = pd.to_numeric(df['Total Orders'], errors='coerce')

# Assuming there is a column with order numbers or IDs that represents the chronological order
df['Order Chronology'] = range(1, len(df) + 1)

order_chronology_stats = df.groupby('Order Chronology').agg({'Total Spent': 'mean', 'Total Orders': 'mean'}).reset_index()

# Plotting the trends
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(order_chronology_stats['Order Chronology'], order_chronology_stats['Total Spent'], marker='o')
plt.title('Average Total Spent Over Time')
plt.xlabel('Order Chronology')
plt.ylabel('Total Spent')

plt.subplot(2, 1, 2)
plt.plot(order_chronology_stats['Order Chronology'], order_chronology_stats['Total Orders'], marker='o')
plt.title('Average Total Orders Over Time')
plt.xlabel('Order Chronology')
plt.ylabel('Total Orders')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Read the CSV file
df = pd.read_csv('customers_export.csv')

# Select relevant columns for analysis (assuming 'Total Spent' is the target variable)
data = df[['Total Spent']]

# Generate a synthetic timestamp column for demonstration purposes
data['Timestamp'] = pd.date_range(start='2022-01-01', periods=len(data), freq='D')

# Set the 'Timestamp' column as the index
data = data.set_index('Timestamp')

# Drop rows with missing values
data = data.dropna()

# Extend timestamp range for future predictions
num_future_periods = 30  # Adjust this based on how far into the future you want to forecast
future_timestamps = pd.date_range(start=data.index[-1] + pd.DateOffset(1), periods=num_future_periods, freq='D')

# Refit the model with the full data
order = (1, 1, 1)  # Autoregressive, Differencing, Moving average order
seasonal_order = (1, 1, 1, 12)

model = SARIMAX(data, order=order, seasonal_order=seasonal_order)
model_fit = model.fit(disp=False)

# Make future predictions
future_predictions = model_fit.get_forecast(steps=num_future_periods)
future_predicted_mean = future_predictions.predicted_mean

# Plot daily predictions
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.xticks(rotation=45)
plt.plot(future_timestamps, future_predicted_mean, color='orange', label='Daily Predictions')
plt.xlabel('Timestamp')
plt.ylabel('Total Spent')
plt.title('SARIMA Model - Daily Sales Forecasting')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Resample for weekly predictions
future_predictions_weekly = future_predicted_mean.resample('W').sum()

# Plot weekly predictions
plt.subplot(132)
plt.xticks(rotation=45)
plt.plot(future_predictions_weekly.index, future_predictions_weekly, color='green', label='Weekly Predictions')
plt.xlabel('Timestamp')
plt.ylabel('Total Spent')
plt.title('SARIMA Model - Weekly Sales Forecasting')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Resample for monthly predictions
future_predictions_monthly = future_predicted_mean.resample('M').sum()

# Plot monthly predictions
plt.subplot(133)
plt.xticks(rotation=45)
plt.plot(future_predictions_monthly.index, future_predictions_monthly, color='blue', label='Monthly Predictions')
plt.xlabel('Timestamp')
plt.ylabel('Total Spent')
plt.title('SARIMA Model - Monthly Sales Forecasting')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv('orders_export_1.csv')

# Convert 'Created at' to datetime
df['Created at'] = pd.to_datetime(df['Created at'])

# Extract month and year information
df['MonthYear'] = df['Created at'].dt.to_period('M')

# Group by month-year and calculate total sales
monthly_sales = df.groupby('MonthYear')['Total'].sum()

# Visualize seasonal trends
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Total Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales (INR)')
plt.grid(True)
plt.show()

# Identify peak months
peak_months = monthly_sales.idxmax()

# Display insights
print(f"The peak sales month is {peak_months}.")
print("Adjustments and insights:")
if peak_months.month in [12, 1, 2]:
    print("Consider offering special promotions or discounts during the holiday season.")
elif peak_months.month in [6, 7, 8]:
    print("Emphasize summer-themed products and promotions.")
else:
    print("Explore marketing strategies to boost sales during off-peak months.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

df = pd.read_csv('orders_export_1.csv')

# Select relevant features and target variable
features = ['Subtotal', 'Shipping', 'Taxes', 'Discount Amount']
target = 'Total'

# Create feature matrix (X) and target vector (y)
X = df[features]
y = df[target]

# Impute missing values in both features and target
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
y_imputed = imputer.fit_transform(y.values.reshape(-1,    1)).ravel()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.2, random_state=42)

# Initialize the Random Forest Regressor
model = RandomForestRegressor(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, predictions)
print(f'Mean Absolute Error: {mae}')

# Plot the actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions, alpha=0.5)
plt.title('Actual vs Predicted Total Values')
plt.xlabel('Actual Total')
plt.ylabel('Predicted Total')
plt.show()

# Plot the trend of actual and predicted values
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual Total', marker='o')
plt.plot(predictions, label='Predicted Total', marker='o')
plt.title('Trend of Actual and Predicted Total Values')
plt.xlabel('Data Points')
plt.ylabel('Total Value')
plt.legend()
plt.show()
```
